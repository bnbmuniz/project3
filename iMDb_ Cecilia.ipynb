```python
# Genres --> div class ipc-chip-list GenresAndPlot__GenresChipList-sc-cum89p-4 cDpOeC
#                      ipc-chip-list GenresAndPlot__GenresChipList-sc-cum89p-4 cDpOeC
response = requests.get(url_list_mov[79])
soup = BeautifulSoup(response.content)
result = soup.find('div', attrs={'class':
                                  re.compile(r'^ipc-chip-list GenresAndPlot__')})

genres = [gen.text for gen in result.find_all('a')]
genres
```

```python
# For each of the 250 movies, create a dict with title, rating, director and duration
movies = []
c =0
for url in url_list_mov[80:200]:
    print(c, url)
    
    #     try:
    response = requests.get(url)
    soup = BeautifulSoup(response.content)
    # add a random time delay
    time.sleep(random.randint(6, 13))
    
    # get the title
    result = soup.find('h1', attrs={'class': lambda e: e.startswith('TitleHeader__TitleTex') if e else False})
    title = result.text

    # get the director
    result = soup.find('a', attrs={'class': 'ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link'})
    director = result.text

    # get the duration
    result = soup.find_all('ul', attrs={'class': 'ipc-inline-list ipc-inline-list--show-dividers TitleBlockMetaData__MetaDataList-sc-12ein40-0 dxizHm baseAlt'})
    hour = result[0].find_all('li')[2].text           
    duration = re.sub('[A-z]', '', hour).replace(' ',':')
    
    # get the genres list
    result = soup.find('div', attrs={'class':
                                  re.compile(r'^ipc-chip-list GenresAndPlot__')})

    genres = [gen.text for gen in result.find_all('a')]              
    
    # get the rating
    result = soup.find('span', attrs={'class':'AggregateRatingButton__RatingScore-sc-1ll29m0-1 iTLWoV'})        
    rating = result.text

    movies.append({'title': title, 
                   'director': director, 
                   'duration': duration, 
                   'rating': rating,
                   'genre': genres
                  })
#     except:
#         pass
    c = c+1
movies   
```
