```python
imdb_movi = pd.read_csv('imdb_movies.csv', index_col=False).drop(['Unnamed: 0'], axis=1)
imdb_show = pd.read_csv('imdb_shows.csv', index_col=False).drop(['Unnamed: 0'], axis=1)
# mov.columns
mov_not_cl = pd.read_csv('imdb_movies_not_cleaned.csv', index_col=False).drop(['Unnamed: 0'], axis=1)
```

```python
# Movies
print('imdb_movi', imdb_movi.shape)
print(len(imdb_movi['title'].unique()))
```

```python
# Drop the duplicate
imdb_movi['title'].duplicated()

imdb_movi[imdb_movi['title'].duplicated()]
# imdb_movi
```

```python
# Shows
print('imdb_show', imdb_show.shape)
print(len(imdb_show['title'].unique()))
```

```python
# Drop the duplicate
imdb_show['title'].duplicated()

imdb_show[imdb_show['title'].duplicated()]
imdb_show[imdb_show['title'] == 'Battlestar Galactica']

imdb_show = imdb_show.drop_duplicates(subset=['title'])
imdb_show
```

```python
# Drop the duplicate
mov_not_cl['title'].duplicated()

mov_not_cl[mov_not_cl['title'].duplicated()]
# mov_not_cl[mov_not_cl['title'] == 'Battlestar Galactica']

mov_not_cl = mov_not_cl.drop_duplicates(subset=['title'])
# mov_not_cl
print('mov_not_cl', mov_not_cl.shape)
print(len(mov_not_cl['title'].unique()))
# mov_not_cl
```

```python
# Add a column for movie or show
mov_not_cl['type'] = 'movie'
imdb_show['type'] = 'show'
mov_not_cl
```

```python
# Concat the two dfs
```

```python
imdb = pd.concat([mov_not_cl, imdb_show], axis=0)

imdb#.shape
```
